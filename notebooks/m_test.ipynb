```python
from cipo import graphics as mpc

# --- CONFIGURAÇÕES DA ANÁLISE ---
DATA_OBSERVACAO = '2025-10-09'

# Filtro de Observação
ALTITUDE_MINIMA = 25
DURACAO_MINIMA = 30

# Dicionário com as páginas que o script pode analisar
TARGET_PAGES = {
    "1": {"name": "NEO", "url": "https://minorplanetcenter.net/iau/NEO/toconfirm_tabular.html"},
    "2": {"name": "PCC", "url": "https://minorplanetcenter.net/iau/NEO/pccp_tabular.html"}
}

# --- FLUXO PRINCIPAL DO PROGRAMA ---

def main():
    """
    Orquestra a análise completa: busca, processa, calcula, filtra e plota.
    """
    # 1. Obter o código do observatório do usuário
    observatory_code = input("Digite o código IAU do observatório (ex: Y28, 500): ")
    
    # 2. Obter as coordenadas para o código fornecido
    coords = mpc.get_observatory_coords(observatory_code)
    if not coords:
        print("Não foi possível continuar sem as coordenadas do observatório.")
        return
    
    lat, lon = coords

    # 3. Menu para escolher a página
    while True:
        print("\n" + "="*40 + "\n  SELECIONE A PÁGINA PARA ANALISAR\n" + "="*40)
        for key, page_info in TARGET_PAGES.items():
            print(f"{key}. {page_info['name']} Confirmation")
        print("0. Sair")
        choice = input("\nDigite sua opção: ")

        if choice == '0':
            print("Encerrando o programa.")
            break
        if choice not in TARGET_PAGES:
            print("Opção inválida, tente novamente.")
            continue

        target_info = TARGET_PAGES[choice]
        
        # 4. Executar o fluxo de análise
        texto_pagina = mpc.fetch_data(target_info['url'], observatory_code)

        if texto_pagina:
            dataframes = mpc.process_data(texto_pagina)
            if dataframes:
                resultados_brutos, intervalo = mpc.calcular_altitudes_para_objetos(
                    dataframes, DATA_OBSERVACAO, lat, lon
                )
                resultados_filtrados = mpc.filtrar_objetos_observaveis(
                    resultados_brutos, ALTITUDE_MINIMA, DURACAO_MINIMA
                )
                mpc.plotar_grafico_altitude(
                    resultados_filtrados, intervalo,
                    {"data": DATA_OBSERVACAO, "lat": lat, "lon": lon},
                    target_info['name'],
                    ALTITUDE_MINIMA, DURACAO_MINIMA
                )

if __name__ == "__main__":
    main()
```
